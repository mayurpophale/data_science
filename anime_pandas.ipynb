```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"/content/anime.csv")
```

```python
df.head()
```

```python
df.loc[2]['Title']
```

```python
def extract_episodas(txt):
  data = ''
  check = False
  for i in txt:
    if i == ')':
      check = False
      return data
    if check == True:
      data +=i
    if i == '(':
      check = True


df["eps"]=df["Title"].apply(extract_episodas)
```

```python
df.head()
```

```python
#make a new column for episode count
df['eps'] = df['eps'].str.replace(" eps",'').astype(int)
```

```python
df.head()
```

```python
#make a new column for time stamp
df['Title'][0]
```

```python
def extract_time(txt):
  check = False
  data = ''
  for i in range(len(txt)):
    if txt[i] == ')':
      for j in range(i+1,i+20):
        data+=txt[j]
      return data
df["Time Stamp"]=df["Title"].apply(extract_time)
```

```python
df.head()
```

```python
dates = df['Time Stamp'].str.split(' - ', expand=True)

start = pd.to_datetime(dates[0], format='%b %Y', errors='coerce')

end = dates[1].replace('Present', pd.Timestamp.today().strftime('%b %Y'))
end = pd.to_datetime(end, format='%b %Y', errors='coerce')

df['Months'] = (
    (end.dt.year - start.dt.year) * 12
    + (end.dt.month - start.dt.month)
    + 1
)
```
