```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=sns.load_dataset('tips')
df
#distribution plots

#1displot or histogram
#2jointplot
#3pairplot
```

```python
df['size'].unique()
```

```python
#displot or histogram
sns.displot(df['total_bill'],bins=20,kde = True)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
#pairplot
# sns.pairplot(df,hue='sex')
sns.pairplot(df,hue='size',palette='rainbow')
```

**Categorical data plots**


1.   boxplot
2.   violinplot
3.   striplot
4.    swarmplot
5.  barplot
6.  countplot





```python
#countplot
df = sns.load_dataset('tips')
sns.countplot(x= df['sex'],hue=df['smoker'])
```

```python
#barplot
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

```python
#boxplot
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

```python
#violinplot
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
```

```python
#striplot
sns.stripplot(x='day',y='tip',data=df,jitter=True,hue='sex',palette='rainbow')
```

```python
#swarmplot
sns.swarmplot(x='day',y='tip',data=df,hue='sex',palette='rainbow')
```

```python
#swarmplot and violinplot combination
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
sns.swarmplot(x='day',y='tip',data=df,color='black',size=3)
```

**Matrixplot**

```python
fl = sns.load_dataset('flights')
fl
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr = tips[['total_bill','tip','size']]
tipscorr.corr()
```

```python
#Heatmap
sns.heatmap(tipscorr.corr(),annot=True,cmap='rainbow')
```

```python
#cluster map
sns.clustermap(tipscorr.corr(),annot=True,cmap='rainbow')
```

```python
#pivottable
pvt = fl.pivot_table(values='passengers',index='month',columns='year')
pvt
```

```python
sns.heatmap(pvt,cmap='magma',linecolor='white',linewidth=1)
```
