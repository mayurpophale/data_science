```python
#basic
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0,11,5)
y =x**2
plt.plot(x,y)
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')


plt.subplot(2,2,2)
plt.plot(y,x,'b')


plt.subplot(2,2,3)
plt.plot(x,y,'g')


plt.subplot(2,2,4)
plt.plot(y,x,'y')
```

```python
#object oriented way:- basically full control of the sheet
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#type of plots
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.bar(x,y)
```

```python
fig = plt.figure()
ax= fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("sample.png")#for saving the plot
```
